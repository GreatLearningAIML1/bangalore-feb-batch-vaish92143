```python
## Q1
import pandas as pd
import numpy as np
df = pd.read_csv('Laliga.csv')
df.replace('-',0,inplace=True)
df.head(10)
```

```python
## Q2
df[(df["Unnamed: 17"] > '1930') & (df["Unnamed: 17"] < '1980')]
```

```python
## Q3
df['Unnamed: 3']=pd.to_numeric(df['Unnamed: 3'],errors='coerce')
df.nlargest(5,'Unnamed: 3')
```

```python
# Q4 
def goal_diff_count():
    df['Unnamed: 8']=pd.to_numeric(df['Unnamed: 8'],errors='coerce')
    df['Unnamed: 9']=pd.to_numeric(df['Unnamed: 9'],errors='coerce')
    df['goal_diff_count'] = df['Unnamed: 8'] - df['Unnamed: 9']
    df.head()
    print ("Maximum goal difference is \n",df['goal_diff_count'].max())
    print ("TEAM WITH MAXIMUM GOAL DIFFERENCE IS..........\n",df.loc[df['goal_diff_count'] == df['goal_diff_count'].max(),'Unnamed: 1'],"\n")
    print ("Minimum goal difference is \n",df['goal_diff_count'].min())
    print ("TEAM WITH MINIMUM GOAL DIFFERENCE IS......... \n",df.loc[df['goal_diff_count'] == df['goal_diff_count'].min(),'Unnamed: 1'])
    
goal_diff_count()
```

```python
## Q5
df['Unnamed: 4']=pd.to_numeric(df['Unnamed: 4'],errors='coerce')
df['Unnamed: 5']=pd.to_numeric(df['Unnamed: 5'],errors='coerce')
df['Winning Percent'] = (df['Unnamed: 5']/df['Unnamed: 4'])*100
df.nlargest(5,'Winning Percent')
```

```python
## Q6
df['Unnamed: 19']=pd.to_numeric(df['Unnamed: 19'],errors='coerce')
df.groupby(['Unnamed: 19']).sum().drop(['Unnamed: 4', 'Unnamed: 5','Unnamed: 8','Unnamed: 9', 'Winning Percent','goal_diff_count'], axis=1)
```
